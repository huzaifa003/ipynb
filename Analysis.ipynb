```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Enable inline plotting
%matplotlib inline

# Load the dataset from your file (adjust the file path if needed)
file_path = "/content/Expanded_Master_Thesis_Survey_438.xlsx"
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Clean column names by stripping whitespace
df.columns = df.columns.str.strip()
# df.dropna(inplace=True)

# (Optional) Display the first few rows to verify data load
df.head()
```

```python
# Create a contingency table
age_medium_ct = pd.crosstab(df["Please select your age range"], df["What is your preferred medium of banking?"])
print("Contingency Table (Age vs Preferred Medium):\n", age_medium_ct)

# Perform Chi-square test
chi2, p, dof, ex = chi2_contingency(age_medium_ct)
print(f"\nChi-square test result: chi2 = {chi2:.2f}, p-value = {p:.4f}")

# Visualization: Stacked Bar Chart
age_medium_ct.plot(kind='bar', stacked=True, figsize=(8,6))
plt.title("Age Group vs. Preferred Medium of Banking")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.legend(title="Preferred Medium")
plt.show()
```

```python
from collections import Counter

def count_services(series):
    counts = Counter()
    for response in series.dropna():
        items = [item.strip() for item in response.split(';')]
        counts.update(items)
    return pd.Series(counts).sort_values(ascending=False)

# Count frequencies for in-person services
inperson_counts = count_services(df["For which services do you prefer to visit the bank in person?\n(You can select multiple options)"])
print("In-person Services Frequency:\n", inperson_counts)

# Count frequencies for digital banking services
digital_counts = count_services(df["For which services do you prefer using digital banking?\n(You can select multiple options)"])
print("\nDigital Banking Services Frequency:\n", digital_counts)

# Visualization: Side-by-side Bar Charts
fig, ax = plt.subplots(1, 2, figsize=(16,6))
inperson_counts.plot(kind='bar', ax=ax[0])
ax[0].set_title("Frequency of In-Person Services")
ax[0].set_ylabel("Count")
digital_counts.plot(kind='bar', ax=ax[1])
ax[1].set_title("Frequency of Digital Banking Services")
ax[1].set_ylabel("Count")
plt.show()
```

```python
# Research Question 14: Identify Factors Correlated with Likelihood of Adopting a Digital-Only Bank

# Define the outcome variable
outcome = "How likely are you to use a digital-only bank instead of a conventional bank in the future?"

# Define candidate predictors with their corresponding column names
predictor_columns = {
    "Ease of Use": "How easy is to use your bank's digital services?",
    "Reliability": "How reliable are your bank’s digital services (e.g., transactions, login, app performance)?",
    "Speed": "How fast are the digital banking services (e.g., transfers, payments)?",
    "Convenience": "Compared to branch banking, how convenient do you find digital banking?",
    "Age": "Please select your age range",
    "Gender": "Please select your gender",
    "Location": "Please select your location"
}

# Loop over predictors to compute contingency tables and chi-square tests
results = {}
for factor, col in predictor_columns.items():
    if col == "How reliable are your bank’s digital services (e.g., transactions, login, app performance)?":
      col = "How reliable are your bankâ€™s digital services (e.g., transactions, login, app performance)?"
    ct = pd.crosstab(df[col], df[outcome])
    chi2_val, p_val, dof, expected = chi2_contingency(ct)
    results[factor] = {"chi2": chi2_val, "p": p_val, "dof": dof}
    print(f"\nFactor: {factor}")
    print("Contingency Table:")
    print(ct)
    print(f"Chi-square test: chi2 = {chi2_val:.2f}, p-value = {p_val:.4f}")

# Visualization: Compare Chi-square Statistics Across Factors
factors = list(results.keys())
chi2_vals = [results[f]["chi2"] for f in factors]
p_vals = [results[f]["p"] for f in factors]

plt.figure(figsize=(10,6))
plt.bar(factors, chi2_vals)
plt.title("Chi-square Statistic for Factors vs. Likelihood of Adopting Digital-Only Bank")
plt.xlabel("Factor")
plt.ylabel("Chi-square Statistic")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualization: Compare P-values Across Factors
plt.figure(figsize=(10,6))
plt.bar(factors, p_vals)
plt.title("P-values for Factors vs. Likelihood of Adopting Digital-Only Bank")
plt.xlabel("Factor")
plt.ylabel("P-value")
plt.axhline(y=0.05, color='red', linestyle='--', label='0.05 threshold')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
